Filtering in data frame using query
1. query always returns a copy of the data frame
2. Column names become variables
3. String values must be in quotes
4. Use backticks for column names with spaces or special characters
5. You can use @ to reference Python variables

```python
df.query("IMDb > 6")
```

```python
df.query("IMDb > 6 and Year==2019")
```

```python
df.query("Actor=='Shah Rukh Khan'")
```

```python
df.query("IMDb>5 and Genre=='Action' | Genre=='Comedy' and Year>2017")
```
