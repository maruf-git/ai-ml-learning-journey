```python
import pandas as pd
import numpy as np
```

```python
# 1d python list and 1d numpy array
print([1,2,3,4,5])
print(type([1,2,3,4,5]))
arr=np.array([1,2,3,4,5])
print(arr)
print(type(arr))
arr.shape

# 2d python list and , 2d numpy array
print([[1,2,3],[4,5,6]])
print(type([[1,2,3],[4,5,6]]))
arr2=np.array([[1,2,3],[4,5,6]])
print(arr2)
print(type(arr2))
```

```python
# creating pandas series using numpy 1d  array
s = pd.Series(arr)
print(s)
print(type(s))

# creating pandas DataFrame using numpy 2d array
s2 = pd.DataFrame(arr2)
print(s2)
print(type(s2))
```

series with cutom index

```python
# s = pd.Series(arr,index=["a","b","c","d","e"])
s = pd.Series([10,20,30,40,50],index=["a","b","c","d","e"])
print(s)
print(type(s))
```

```python
# accessing series elements
print(s["a"])
print(s["e"])
```
