Vetorization

```python
# importing numpy
import numpy as np
```

```python
# vectorization
arr = np.array([1,2,3,4,5])
print(arr)
result = arr ** 2 #vectorized operation
print("result array: ",result)
```

What is Broadcasting?

Broadcasting is NumPy’s way of handling operations between arrays of different shapes without making unnecessary copies.

The Rule of Broadcasting:

For two arrays to be compatible:

If dimensions match → ✅ works directly.

If one dimension is 1 → ✅ NumPy stretches it.

Otherwise → ❌ error.

```python
# Broadcasting
result = result + 10 # result is an array and 10 is an integer
result
```

```python
# Broadcasting in 2d array
arr1 = np.array([[1,2,3],
                 [4,5,6]])
arr2 = np.array([1,2,3])
result = arr1+arr2
print(result)
```

Normalization using Broadcasting

```python
data = np.array([[10,20,30],
                [15,25,35],
                [20, 30,40],
                [25,35,45],
                [30,40,50]])
print(data)
print(data.shape)
mean = data.mean(axis=0) #mean
print("mean is: ",mean)
std = data.std(axis=0) #standard deviation
print(std)
```

```python
# normalizing the data
normalized_data = (data-mean)/std
print(normalized_data)
```
