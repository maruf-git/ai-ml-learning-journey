Build in mathematical functions in numpy

```python
import numpy as np
```

```python
data = np.array([[10,20,30],
                [15,25,35],
                [20, 30,40],
                [25,35,45],
                [30,40,50]])
mean = np.mean(data,axis=0) # this is built in numpy mean function
print(mean)

mean2 = data.mean(axis=0) #this is numpy array object method
print(mean2)
```
