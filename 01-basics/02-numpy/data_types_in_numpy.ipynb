Numpy Array data types

```python
# numpy arrays are homogenous
# they can store only one type of data at a time like c and c++

# int32, int64: for integer types
# float32,float64: for floating points types
# bool: Boolean data type
# complex64, complex128: Complex number types
# object: For storing objects(e.g. Python objects, strings)
```

```python
import numpy as np
arr = np.array([1,2,3,4,5],dtype="float") #creating float type array
print(arr.dtype)
arr
```

```python
# type casting in numpy
arr.dtype
arr = arr.astype("int")
print(arr.dtype)
arr
```

```python
# string
arr = arr.astype("str")
print(arr.dtype)
print(arr)
```

```python
# object/dictionary
arr = np.array([{'a':1},{'b':2}])
print(arr.dtype)
print(arr)
```
