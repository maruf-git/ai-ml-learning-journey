# Class and Object

```python
# declaring a class
class Phone:
    category="Electronics" # class attribute
    
    # constructor
    def __init__(self,model,battery, camera,battery_percentage =100): # object/instance attribute
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage 
        
    # method
    def charge(self, hour):
        self.charge  +=hour
```

```python
# creating object
apple = Phone("Iphone 17",3000,20)
print(apple.category)
apple.category="super electronics"
print(apple.category)
print(apple.camera)
print(apple.battery_percentage)

```

## Inheritance

```python
#  child / derived class
class SmartPhone(Phone):
    # constructor
    def __init__(self,model,battery, camera, processor):
        super().__init__(model,battery,camera)
        self.processor = processor    
```

```python
pro = SmartPhone("x",5000,100,"Snapdragon")
print(pro)
print(pro.model)
print(pro.battery)
print(pro.camera)
print(pro.processor)
```

# Constructor
python does not support constructor overloading

```python
class Student:
    counter=0
    
    # default constructor:
    def __init__(self):
        print("This is default constructor")
    
    # parameterized constructor
    def __init__(self,name,marks):
        print("new student is adding to the database")
        self.name = name
        self.marks = marks
        Student.counter +=1
        self.id = Student.counter
```

```python
s1 = Student("maruf",90)
print(s1.counter,s1.id,s1.name,s1.marks)
```

```python
s2 = Student("rifat",90)
print(s2.counter,s2.id,s2.name,s2.marks)
```

# Class and Instance Attribute

```python
class Student:
    name = "anonymous" # class attribute
    

    # parameterized constructor
    def __init__(self,name,marks): # object attribute
        print("new student is adding to the database")
        self.name = name
        self.marks = marks
        
    # method
    def welcome(self):
        print("Welcome Student",self.name)
     
    
```

```python
s1 = Student("maruf",90)
print(s1.name,Student.name)
print(s1.welcome())
```

# Static method
methods that dont use the self parameter(work at class level)

```python
class Student:
    name = "anonymous" # class attribute
    
    # static method
    @staticmethod
    def welcome():
        print("Welcome Student")

    # parameterized constructor
    def __init__(self,name,marks): # object attribute
        print("new student is adding to the database")
        self.name = name
        self.marks = marks
        
  
     
    
```

```python
s1 = Student("maruf",90)
print(s1.welcome())
```

# Abstraction
hinding the implementation details of a class and only showing the essential features to the user

```python
class Car:
    def __init__(self):
        self.acc = False
        self.brk = False
        self.clutch = False
    def start(self):
        self.cluth = True
        self.acc = True
        print("Car Started")
```

```python
car1 = Car()
car1.start()
```

# Del Keyword
Used to delete object properties or object itself

```python
class Student:
    def __init__(self,name):
        self.name = name
```

```python
s1 = Student("maruf")
print(s1.name)
print(s1)
del s1.name
print(s1.name)
```
